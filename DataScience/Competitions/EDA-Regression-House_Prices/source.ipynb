```python
def concatenate_datasets(train: pd.DataFrame, test: pd.DataFrame) -> pd.DataFrame:
    """Concatenates train and test datasets, adding a new column to identify the origin of each row.

    Args:
        train (pd.DataFrame): The training dataset.
        test (pd.DataFrame): The testing dataset.

    Returns:
        pd.DataFrame: The concatenated dataset.
    """
    train["Group"] = "train"
    test["Group"] = "test"
    
    return pd.concat([train, test], ignore_index=True)
```

```python
def find_data_types(df: pd.DataFrame):
    """
    Separates the columns of a DataFrame into numerical, and categorical based on their data types and unique values.

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        num_cols (List[str]): List of numerical column names.
        cat_cols (List[str]): List of categorical column names.
    """
    num_cols = df.select_dtypes("number").columns.to_list()
    cat_cols = df.select_dtypes(include=["object", "bool"]).columns.to_list()
    
    return num_cols, cat_cols
```

```python
def plot_histograms_with_skewness(df, columns, nrows, ncols, title):
    """
    Plots histograms for a list of columns from a dataset, annotating each plot with the skewness of the data.

    Args:
        dataset (pd.DataFrame): The input DataFrame.
        columns (list): List of column names to plot.
        nrows (int): Number of rows in the subplot grid.
        ncols (int): Number of columns in the subplot grid.
        title (str): Title for the entire plot.

    Returns:
        None
    """
    fig, axs = plt.subplots(nrows, ncols, figsize=(24, 32))
    fig.suptitle(title, y=1, size=20)
    axs = axs.ravel()

    for ax, column in zip(axs, columns):
        skewness = round(df[column].skew(skipna=True), 2)
        sns.histplot(df[column], ax=ax, bins=10, fill=True, alpha=0.5, linewidth=0)
        ax.set_title(f"{column}, skewness is {skewness}")

    plt.tight_layout()
    plt.show()
```

<div style="font-size:14px; font-family:Verdana; border-radius:10px; border: #4E5672 solid; padding: 15px; background-color: #CCFFCC; text-align:left; color: #4E5672">


<h3 align="left"><font color='#4E5672'>üí° Comment:</font></h3>

* For non-numeric columns in a Pandas DataFrame, calculate the following:

  * Percentage of missing values: This is the percentage of values in the column that are missing, from 0 to 100.
  
  * Number of unique values: This is the number of different values in the column, excluding missing values. For categorical columns, this is the same as the number of categories.
  
  * Percentage of each value: This is the percentage of each unique value in the column, sorted by decreasing percentage. Missing values are excluded from this calculation.

* To calculate the percentage of each value, you can use a dictionary to store the results, where the key is the value and the value is the percentage. For categorical columns, the method is the same.

* Save the results to a Pandas DataFrame called cat_col_info_df, with the following columns:

  * Column name: The name of the non-numeric column in the original DataFrame (raw_df)
  
  * missing_ratio: The percentage of missing values in the column
  
  * num_values: The number of unique values in the column, excluding missing values
  
  * value_ratios: A dictionary of the percentage of each unique value in the column, sorted by decreasing percentage

* To improve readability, round all values to one decimal place using the .round(1) method.
  
</div>

```python
cat_cols_info = df.select_dtypes(include=['object', 'bool'])
cat_cols_info.drop('Group', axis=1, inplace=True)

def missing_ratio(s):
    return (s.isna().mean() * 100)

def num_values(s):
    s = s.astype(str).str.split(';')
    s = s.explode()
    return len(s.value_counts())

def value_ratios(s):
    s = s.astype(str).str.split(';')
    s = s.explode()
    totalCount = (~s.isna()).sum()
    return ((s.value_counts()/totalCount*100).round(3)).to_dict()

cat_cols_info_df = cat_cols_info.agg([missing_ratio, num_values, value_ratios])
cat_cols_info_df
```

```python
def plot_categorical_means(df, cat_cols, target):
    """
    Plots bar plots for each categorical column against the target column, 
    with the bars sorted by the mean target value for each category.

    Args:
        df (pd.DataFrame): The input DataFrame.
        cat_cols (list): List of categorical column names to plot.
        target (str): The target column name.

    Returns:
        None
    """
    nrows = len(cat_cols) // 3 + 1
    ncols = 3
    fig, axs = plt.subplots(nrows, ncols, figsize=(len(cat_cols)//3, 3))
    axs = axs.ravel()

    for idx, column in enumerate(cat_cols):
        # Calculate mean target value for each category
        data = df[df["group"] == "train"].groupby(column)[target].mean()
        data = data.reset_index().sort_values(by=target)

        ax = axs[idx]
        sns.barplot(y=column, x=target, data=data, color='grey', ax=ax)

        # Annotate bars with the mean target value
        for p, count in enumerate(data[target].values):
            ax.text(count + 10, p + 0.05 / len(data), f"{int(count // 1000)}k", color='black', fontsize=11)

        ax.set_title(f"{column} and {target}")
        ax.set_xticks([], minor=False)
        ax.set_box(False)

    plt.tight_layout()
```

<div style="font-size:14px; font-family:Verdana; border-radius:10px; border: #4E5672 solid; padding: 15px; background-color: #CCFFCC; text-align:left; color: #4E5672">


<h3 align="left"><font color='#4E5672'>üí° Comment:</font></h3>

* Once the basic statistical properties of the data have been established, the next step is to identify features with a significant number of missing values. These features are not conducive to the analysis process and should be eliminated from the dataset.

* The threshold for determining "significant" missingness can be tailored to specific objectives. As a general rule, if a feature exhibits a missing value percentage exceeding 75%, it is removed from the dataframe, resulting in an updated dataframe.
  
</div>


```python
def drop_missing_features(df: pd.DataFrame, missing_lst: dict = dict(num_cols_info_df.iloc[0]), threshold: float = 75.0) -> pd.DataFrame:
    """Drop missing features from Pandas dataframe base on given threshold.

    Args:
        df (pd.DataFrame): Input dataframe for processing.
        missing_lst (dict, optional): Dict missing value rate which calculated from previous stage. Defaults to dict(num_col_info_df.iloc[0]).
        threshold (float, optional): User threshold for dropping. Defaults to 75.0.

    Returns:
        pd.DataFrame: Updated dataframe
    """
    if (df is None) or (missing_lst is None) or (threshold is None):
        print(f"[LOG] Invalid.")
        raise ValueError
    
    # Create a copy of the dataframe to avoid changing the original
    df_cp = df.copy()
    
    # List to maintain the columns to drop
    cols_to_trim = []
    
    for key, value in missing_lst.items():
        if float(value) > threshold:
            cols_to_trim.append(key)
            
    if len(cols_to_trim) > 0:
        # drop the columns identified using the dataframe drop() method
        df_cp = df_cp.drop(columns=cols_to_trim)
        print("[LOG] Dropped column(s): " + " ".join(cols_to_trim))
    else:
        print("[LOG] Have no column(s) to trim.")
        
    return df_cp
```

<h3 align="left"><font color='#4E5672'>üí° Comment:</font></h3>

* Despite eliminating features with a substantial number of missing values, missing values persist in our dataframe. Consequently, we must impute these missing values to render the data suitable for analysis.

```python
def filling_missing_value(df, missing_num_cols, missing_cat_cols):
    """
    Filling missing value for numerical and categorical columns.

    Args:
        df (pd.DataFrame): The input DataFrame.
        missing_num_cols (list): List of numerical column names to fill.
        missing_cat_cols (list): List of categorical column names to fill.

    Returns:
        None
    """
    # Fill missing values for numerical columns
    for column in missing_num_cols:
        df[column].fillna(0, inplace=True)

    # Add "None" as a category and fill missing values for categorical columns
    for column in missing_cat_cols:
        df[column].fillna("None", inplace=True)

    return df
```

```python
def detect_outliers(df, cols, handle_outliers=False, print_outliers=False, q1=0.25, q3=0.75):
    """
    Detect and optionally handle outliers in a DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.
        cols (list): List of column names to check.
        handle_outliers (bool): Whether to handle outliers by replacing them with the upper/lower limit.
        print_outliers (bool): Whether to print the outliers.
        q1 (float): The lower quartile.
        q3 (float): The upper quartile.

    Returns:
        df (pd.DataFrame): The DataFrame with outliers handled (if handle_outliers is True).
        temp (pd.DataFrame): The DataFrame with information about the outliers (if print_outliers is True).
    """
    outlier_types = {
        (True, True): "Min-Max-Outlier",
        (True, False): "Max-Outlier",
        (False, True): "Min-Outlier",
        (False, False): "No"
    }

    temp = pd.DataFrame(index=cols, columns=["Min", "Low_Limit", "Mean", "Median", "Up_Limit", "Max", "Outliner"])

    for col in cols:
        q1_val = df[col].quantile(q1)
        q3_val = df[col].quantile(q3)
        IQR = q3_val - q1_val
        up = q3_val + 1.5 * IQR
        low = q1_val - 1.5 * IQR
        temp.loc[col] = [round(df[col].min()), round(low), round(df[col].mean()), round(df[col].median()), up, df[col].max(), outlier_types[(df[col].max() > up, low > df[col].min())]]

        if handle_outliers:
            df.loc[df[col] > up, col] = round(up - 1)
            df.loc[df[col] < low, col] = round(low - 1)

    return (df if handle_outliers else temp) if print_outliers else df
```

> <span style='font-size:15px; font-family:Arial; color: #87CEFA;'><b>
Does the most frequent value in the feature represent more than a given percentage of the total values?</b></span>


```python
def drop_nonsense_features(df, columns, percentage):
    """
    Drop features from a DataFrame if the most common value in the feature 
    makes up more than a specified percentage of the total values.

    Args:
        df (pd.DataFrame): The input DataFrame.
        columns (list): List of column names to check.
        percentage (float): The percentage threshold for dropping a feature.

    Returns:
        df (pd.DataFrame): The DataFrame with nonsense features dropped.
        new_cat_cols (list): List of remaining categorical column names.
    """
    new_cat_cols = [col for col in columns if (df[col].value_counts(normalize=True).iloc[0] * 100) <= percentage]
    dropped_cols = list(set(columns) - set(new_cat_cols))

    for col in dropped_cols:
        print(f"Feature {col} is nonsense, dropped")

    df = df.drop(dropped_cols, axis=1)

    return df, new_cat_cols
```

```python
def build_model(model, X_train=X_train, y_train=y_train, X_test=X_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return model, y_pred

def Reg_Models_Evaluation_Metrics(model, X_train, y_train, X_test, y_test, y_pred):
    # Calculate metrics
    r2 = model.score(X_test, y_test)
    cv_r2 = cross_val_score(model, X_train, y_train, cv=10).mean()
    rmse = mean_squared_error(y_test, y_pred, squared=False)

    # Calculate adjusted R-squared
    n = X_test.shape[0]
    p = X_test.shape[1]
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

    return r2, adjusted_r2, cv_r2, rmse
```

```python
result= {}
y_preds = {}

for model_name, model_instance in models.items():
    print(f"Training {model_name}...")
    model_instance, y_pred = build_model(model_instance)
    models[model_name] = model_instance
    y_preds[model_name] = y_pred
    
    metrics = Reg_Models_Evaluation_Metrics(model_instance, X_train, y_train, X_test, y_test, y_pred)
    result[model_name] = metrics
```

```python
def objective_rf(trial):
    
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'max_features': trial.suggest_categorical('max_features', ['auto', 'sqrt', 'log2']),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'bootstrap': trial.suggest_categorical('bootstrap', [True, False]),
        'random_state': 42
    }
    
    model_rf = RandomForestRegressor(**param)
    model_rf.fit(X_train, y_train)
    y_pred = model_rf.predict(X_test)
    return mean_squared_error(y_test, y_pred, squared=False)
```

```python
def objective_gbr(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'max_features': trial.suggest_categorical('max_features', ['auto', 'sqrt', 'log2']),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.001, 0.1),
        'random_state': 42
    }
    
    model_gbr = GradientBoostingRegressor(**params)
    model_gbr.fit(X_train, y_train)
    y_pred = model_gbr.predict(X_test)
    return mean_squared_error(y_test, y_pred, squared=False)
```

```python
import optuna
def objective_xg(trial):
    """Define the objective function"""

    params = {
        'booster': trial.suggest_categorical('booster', ['gbtree']),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'max_leaves': trial.suggest_int('max_leaves', 8, 1024),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
        'n_estimators': trial.suggest_int('n_estimators', 400, 1500),
        'min_child_weight': trial.suggest_int('min_child_weight', 10, 20),
        'subsample': trial.suggest_float('subsample', 0.3, 0.9),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.01, 0.5),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.5, 1.0),
        'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.3, 0.8),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1),
        'colsample_bynode': trial.suggest_float('colsample_bynode', 0.01, 0.5),
        "random_state" : trial.suggest_categorical('random_state', [42]),
        'objective': trial.suggest_categorical('objective', ['reg:squarederror']),
        "n_jobs" : trial.suggest_categorical('n_jobs', [-1]),
    }

    model_xgb = XGBRegressor(**params)
    model_xgb.fit(X_train, y_train)
    y_pred = model_xgb.predict(X_test)
    return mean_squared_error(y_test, y_pred, squared=False)
```
