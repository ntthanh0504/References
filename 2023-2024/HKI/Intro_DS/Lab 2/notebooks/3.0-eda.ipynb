## Making your own question 01? (1.5 points)

**A possible question is:** What is the relationship between renewable energy consumption and GDP per capita?

**Answering this question will** help us to
- Understand the relationship between economic development and environmental sustainability.
- Identify countries that are leading the way in adopting renewable energy, and those that are lagging behind.
- Generate insights into how to promote renewable energy adoption in countries with low GDP per capita.

**How we answer this question:** We will compare the average renewable energy consumption and GDP in USD for the largest countries in the world, while controlling for other factors that may influence the relationship, such as population size.


### Pre-processing

```python
import statsmodels.api as sm
```

```python
def get_data(df: pd.DataFrame, selected_columns: list) -> pd.DataFrame:
    """Returns a DataFrame containing the selected columns from the input DataFrame."""
    return df[selected_columns]

def group_by(df: pd.DataFrame, key: str, columns: list, method: str) -> pd.DataFrame:
    """Returns a DataFrame grouped by the specified key and aggregated using the specified method."""
    return df.groupby(key)[columns].agg(method).reset_index().\
        sort_values(by=columns, ascending=False).reset_index(drop=True).\
        rename(columns={column: f'Avg.{column}' for column in columns})

def plot_top_values(df: pd.DataFrame, x: str, y: str, top_n: int, ax: plt.Axes = None) -> None:
    """Plots a bar chart of the top `top_n` values in the specified column `y` of the input DataFrame, grouped by the specified column `x`.

    Args:
        df: The DataFrame to plot.
        x: The column to group by.
        y: The column to plot.
        top_n: The number of top values to plot.
        ax: The optional Axes object to plot on.
    """

    colors = ['grey'] * len(df)
    colors[df[x].idxmax()] = 'red'

    sns.barplot(x=x, y=y, data=df.head(top_n), palette=colors, ax=ax)
```

```python
# set data to be used
selected_columns = ['Country', 'Renewable Energy Consumption (%)', 'GDP in USD', 'Total Population']
df_copy = cleaned_df.copy()
data = df_copy[selected_columns]
data.head()
```

### Analyzing to get the answer

Step 01: Comparing the average renewable energy consumption and GDP in USD for the largest countries in the world

```python
# exclude mean of renewable energy consumption and gdp in usd in each country
avg_consumption_by_country  = group_by(data, key='Country', columns=['Renewable Energy Consumption (%)'], method='mean')
avg_gdp_by_country = group_by(data, key='Country', columns=['GDP in USD'], method='mean')

print("Average of renewable energy consumption in each country")
print(avg_consumption_by_country.head())
print("\nAverage of GDP in each country")
print(avg_gdp_by_country.head())
```

```python
# Plot the top 5 countries with the highest renewable energy consumption
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

ax[0].set_title('Avg. Renewable Energy Consumption', fontsize=18, fontweight='bold', pad=30)
ax[0].set_xlabel('Avg. Renewable Energy Consumption (%)', fontsize=16, labelpad=20)
ax[0].set_ylabel('Country', fontsize=16, labelpad=20)

plot_top_values(avg_consumption_by_country, x='Avg.Renewable Energy Consumption (%)', y='Country', top_n=5, ax=ax[0])

# Plot the top 5 countries with the highest GDP
ax[1].set_title('Avg. GDP in USD', fontsize=18, fontweight='bold', pad=30)
ax[1].set_xlabel('Avg. GDP in USD', fontsize=16, labelpad=20)
ax[1].set_ylabel('Country', fontsize=16, labelpad=20)

plot_top_values(avg_gdp_by_country, x='Avg.GDP in USD', y='Country', top_n=5, ax=ax[1])

# Add a suptitle to the figure
plt.suptitle('Top 5 countries with the highest feature', fontsize=20, fontweight='bold', y=1.05)

# Tighten the layout and show the plot
plt.tight_layout()
plt.show()
```

Step 02: Estimate correlation between renewable energy consumption and GDP in USD, while controllingpopulation size.

```python
# Compute the standardized GDP in USD
gdp_std = (data['GDP in USD'] - data['GDP in USD'].mean()) / data['GDP in USD'].std()
total_population_std = (data['Total Population'] - data['Total Population'].mean()) / data['Total Population'].std()
# Create a new DataFrame with standardized GDP in USD and other features
data_std = pd.DataFrame({
    'Renewable Energy Consumption (%)': data['Renewable Energy Consumption (%)'],
    'GDP in USD (std)': gdp_std,
    'Total Population (std)': total_population_std
})

# Compute the correlation matrix
corr_matrix = data_std[['GDP in USD (std)', 'Renewable Energy Consumption (%)']].corr()

# Plot a scatter plot and correlation matrix
fig, ax = plt.subplots(1, 2, figsize=(18,6))

sns.scatterplot(data=data_std, x='GDP in USD (std)', y='Renewable Energy Consumption (%)', ax=ax[0])
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax[1])

# Set the titles and labels of the plots
plt.suptitle('Renewable Energy Consumption vs. GDP per capita', fontsize=20, fontweight='bold', y=1.05)
ax[0].set_title('Scatter plot', fontsize=16, fontweight='bold', pad=20)
ax[1].set_title('Correlation Matrix', fontsize=16, fontweight='bold', pad=20)

ax[0].set_xlabel('GDP per capita (USD)', fontsize=16, labelpad=20)
ax[0].set_ylabel('Renewable Energy Consumption (%)', fontsize=16, labelpad=20)

plt.tight_layout()
plt.show()
```

```python
# Define the dependent and independent variables
dependent_var = data_std['Renewable Energy Consumption (%)']
independent_vars = data_std[['GDP in USD (std)', 'Total Population (std)']]

# Add a constant term to the independent variables
independent_vars = sm.add_constant(independent_vars)

# Fit the multivariate regression model
model = sm.OLS(dependent_var, independent_vars)
results = model.fit()

# Print the model summary
print(results.summary())
```

Step 03: My observations and thoughts

Differences exist between the top 5 countries with the highest average use of renewable energy and the top 5 countries with the highest average GDP in the USA.

There is a negative connection between the consumption of renewable energy and the GDP in the USA, with a correlation of -0.25. In simple terms, when the percentage of energy consumption from renewables goes up, the GDP in the USA tends to go down.

Let's examine the numbers:

- If the GDP in USD (standardized) increases by one unit, the percentage of renewable energy consumption decreases by 19.81 units.
- If the Total Population (standardized) increases by one unit, the percentage of renewable energy consumption goes up by 16.56 units.

Now, let's look at the P-values for these figures:

- The P-value for Total Population (standardized), GDP in the USA (standardized), and the Intercept is 0.00.
- All of these P-values are statistically significant, as they are less than 0.05.

Therefore, we can conclude that there is a relationship between Total Population (standardized) and GDP in the USA (standardized) with the percentage of renewable energy consumption.

**Answering this question will** help us to learn about population growth and gender trends will help us understand how evenly distributed the global population is and how gender composition is changing.

**How we answer this question:** We will collect data on total population by continent and country over time to show population trends.


### Pre-Processing

```python
selected_cols = ['Country', 'Year', 'Total Population', 'Male Population', 'Female Population']
data = df_copy[selected_cols].copy()
data.head()
```

```python
def get_population_data(data, country=None):
    """Returns the population data of the specified country or total if country is None."""
    if country:
        data = data[data['Country'] == country]
    return data.groupby('Year')[['Total Population', 'Male Population', 'Female Population']].sum().reset_index()

def plot_population_data(data, ax):
    """Plots the population data."""
    melted_data = data.melt(id_vars='Year', var_name='Population', value_name='Count')
    sns.lineplot(x='Year', y='Count', hue='Population', data=melted_data, ax=ax)
```

### Conclusion

Renewable energy consumption is higher than fossil fuel consumption in all African countries.

```python
selected_cols = ['Country', 'Employment in Industry(%)', 'Employment in Agriculture(%)', 'Female Employment in Agriculture(%)', 'Female Employment in Industry(%)', 'Unemployment(%)']
data = df_copy[selected_cols].copy()
data.head()
```

### Analyzing to get answer

```python
data.describe()
```

```python
def hist_plots(df, columns, bins=20, color='grey'):
    fig, ax = plt.subplots(2, 3, figsize=(20, 10))

    for i, column in enumerate(columns):
        sns.histplot(data=df, x=column, bins=bins, kde=True, color=color, ax=ax[i//3, i%3])
        ax[i//3, i%3].set_xlabel(column, fontsize=14, labelpad=20)
        ax[i//3, i%3].set_ylabel('Count', fontsize=14, labelpad=20)

    plt.tight_layout()
    plt.show()
    
hist_plots(data, columns=data.columns[1:].to_list())
```

### Conclusion

Employment in agriculture is likely to be higher than employment in industry in African countries, with the percentage of workers employed in agriculture ranging from 20% to 60% and the percentage of workers employed in industry ranging from 5% to 15%. 

Additionally, women are more likely to be employed in agriculture than in industry, with the percentage of female workers employed in agriculture ranging from 40% to 80%. 

The unemployment rate in all African countries is likely to be between 0% and 5%.
