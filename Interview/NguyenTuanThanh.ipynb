<h3 align="left"><font color='#4E5672'>üìù Description:</font></h3>

* **Objective:** This competition aims to predict the sale prices of houses by analyzing a range of factors that influence their market value.

* **Factors Affecting House Prices:**

  * The value of a house is determined by a complex interplay of various factors, including:
    
  * Location: The location of a property plays a significant role in its price, with desirable neighborhoods and proximity to amenities commanding higher valuations.
    
  * Size and Layout: The size and layout of a house, particularly the number of bedrooms and bathrooms, contribute to its overall appeal and market value.
    
  * Condition and Quality: The condition and quality of construction materials used in the property significantly impact its valuation. Newer or modernized homes with high-quality materials tend to be valued more highly.
    
  * Age and Renovations: The age of the property and any recent renovations or updates are crucial considerations. Newer or renovated homes with modernized features typically command higher prices.
    
  * Outdoor Features: Outdoor features such as a well-maintained yard, a swimming pool, or additional recreational spaces can enhance a property's attractiveness and increase its market value.
    
  * Unique Characteristics: Unique architectural features, special amenities, and energy-efficient elements can also positively impact house prices, setting them apart from similar properties.

  * Garage and Parking: The presence of a garage and sufficient parking space are essential factors that can influence a property's value.

  * Market Conditions: Broader market conditions, including supply and demand, play a significant role in determining house prices. In a seller's market with high demand and low supply, prices tend to rise, while the opposite occurs in a buyer's market.

* **Conclusion:** The interplay of these diverse factors shapes the perceived value of a property in the housing market. Understanding and analyzing these factors is crucial for accurate house price prediction.

Outlier Detection</b></span>

```python
def detectOutliers(df, cols, handle_outliers=False, print_outliers=False, q1=0.25, q3=0.75):
    """
    Detect and optionally handle outliers in a DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.
        cols (list): List of column names to check.
        handle_outliers (bool): Whether to handle outliers by replacing them with the upper/lower limit.
        print_outliers (bool): Whether to print the outliers.
        q1 (float): The lower quartile.
        q3 (float): The upper quartile.

    Returns:
        df (pd.DataFrame): The DataFrame with outliers handled (if handle_outliers is True).
        temp (pd.DataFrame): The DataFrame with information about the outliers (if print_outliers is True).
    """
    outlier_types = {
        (True, True): "Min-Max-Outlier",
        (True, False): "Max-Outlier",
        (False, True): "Min-Outlier",
        (False, False): "No"
    }

    temp = pd.DataFrame(index=cols, columns=["Min", "Low_Limit", "Mean", "Median", "Up_Limit", "Max", "Outliner", "Num_Outliers"])

    for col in cols:
        q1_val = df[col].quantile(q1)
        q3_val = df[col].quantile(q3)
        IQR = q3_val - q1_val
        up = q3_val + 1.5 * IQR
        low = q1_val - 1.5 * IQR
        num_outliers = ((df[col] > up) | (df[col] < low)).sum()
        temp.loc[col] = [round(df[col].min()), round(low), round(df[col].mean()), round(df[col].median()), up, df[col].max(), outlier_types[(df[col].max() > up, low > df[col].min())], num_outliers]

        if handle_outliers:
            df.loc[df[col] > up, col] = round(up - 1)
            df.loc[df[col] < low, col] = round(low - 1)

    return (df if handle_outliers else temp) if print_outliers else df
```

```python
def regModelsEvaluationMetrics(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    
    # Calculate metrics
    r2 = model.score(X_val, y_val)
    kf = KFold(n_splits=5,shuffle=True,random_state=seed)
    cv_r2 = cross_val_score(model, X, y, cv=kf).mean()
    rmse = mean_squared_error(y_val, y_pred, squared=False)

    # Calculate adjusted R-squared
    n = X_val.shape[0]
    p = X_val.shape[1]
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

    return r2, adjusted_r2, cv_r2, rmse

def plotMetrics(result):
    fig, axs = plt.subplots(2, 2, figsize=(16, 10))
    fig.suptitle('Regression Models Evaluation Metrics', y=1, size=20)
    axs = axs.ravel()

    for i, metric in enumerate(['R2', 'ADJUSTED R2', 'CV R2', 'RMSE']):
        values = [item[i] for item in result.values()]
        colors = ['red' if (x == max(values) and metric != 'RMSE') or (x == min(values) and metric == 'RMSE')  else 'grey' for x in values]
        axs[i].bar(result.keys(), values, color='grey')
        axs[i].set_title(metric)
        axs[i].set_ylim(0, 1)
        axs[i].set_yticks([0, 0.25, 0.5, 0.75, 1])

    plt.tight_layout()
```
